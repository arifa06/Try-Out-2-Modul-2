SOAL No. 1

```python
# 'Price'
prc = khd['price']
pd.DataFrame(prc).describe()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
sns.histplot(prc, bins=50, color='blue', label='Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Histogram')
plt.legend(loc=0)
plt.grid(True)
plt.show()
```

```python
# 'Grade'
grd = khd['grade']
pd.DataFrame(grd).describe()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
sns.histplot(grd, bins=10, color='red', label='Grade')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Grade')
plt.legend(loc=0)
plt.grid(True)
plt.show()
```

SOAL No. 2

```python
q1 = khd['sqft_living'].quantile(0.25)
q3 = khd['sqft_living'].quantile(0.75)

iqr = q3-q1

lower = q1-1.5*iqr
upper = q3+1.5*iqr

print('Lower Bound :', lower)
print('Upper Bound :', upper)
```

```python
outliers = khd[(khd['sqft_living']>upper) | (khd['sqft_living']<lower)]
outliers.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=khd, x=khd['sqft_living'])
plt.title('sqft_living Boxplot')
plt.xlabel('sqft_living')
plt.grid(True)
plt.show()
```

```python
cprice = khd[khd['price'] > khd['price'].mean()]
cprice.head()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
sns.barplot(x= cprice['zipcode'].head(5), y=cprice['price'].head(5))
plt.xlabel('Zipcodee')
plt.ylabel('Price')
plt.title('Zipcode with The Highest Price')
plt.grid(True)
plt.show()
```
